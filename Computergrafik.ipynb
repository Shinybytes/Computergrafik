### Aufgabe 1A
Berechnen Sie die Koordinaten der Pixel einer Geraden
mit dem Anfangspunkt (2;3) und dem Endpunkt (11;8).
Geben Sie zus√§tzlich die zugeh√∂rigen Werte der
Entscheidungsvariablen $en$ an.


```python
def bresenham(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    x = x0
    y = y0
    e = 2 * dy - dx
    while x <= x1:
        print(f'({x}, {y}, {e})')
        if e >= 0:
            y += 1
            e -= 2 * dx
        x += 1
        e += 2 * dy

print("Ergebnis 1A:\n(x, y, en)\n ")
bresenham(2, 3, 11, 8)
```

### Aufgabe 3
Anstieg des linearen Teils der Gamma-Funktion f√ºr BT.709

Gegeben sind die Basiswerte der EOTF-BT.709-Gamma-Funktion:  
‚Ä¢ Der Versatz A = 0,099
‚Ä¢ Der Gamma-Exponent: ùö™ = 2,222
‚Ä¢ Um eine Kontinuit√§t der Tangenten im Ber√ºhrungspunkt der beiden Kurven (linearer
Teil und Potenzfunktion) zu gew√§hrleisten, muss ein Schnittpunkt gefunden werden.
‚Ä¢ Berechnen Sie f√ºr den Schnittpunkt (X,X/Œ¶)
‚Ä¢ Die Steigung Œ¶ des linearen Abschnitts
‚Ä¢ Die X-Koordinate des Schnittpunkts (X)


```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

# Gegebene Werte
A = 0.099
Gamma = 2.222

# Funktionen definieren
def linear(x, phi):
    return x / phi

def power(x):
    return ((x + A) ** Gamma) / (1 + A ** Gamma)

# Gleichungssystem definieren
def equations(vars):
    X, phi = vars
    eq1 = linear(X, phi) - power(X)
    eq2 = 1/phi - (Gamma * (X + A) ** (Gamma - 1)) / (1 + A ** Gamma)
    return [eq1, eq2]

# Gleichungssystem l√∂sen
X, phi =  fsolve(equations, (0.1, 1))

# Ausgabe der L√∂sung
print(f"Die Steigung œÜ des linearen Abschnitts betr√§gt: {phi}")
print(f"Die X-Koordinate des Schnittpunkts (X) betr√§gt: {X}")

# Plot erstellen
x = np.linspace(0, 1, 400)
y_linear = linear(x, phi)
y_power = power(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_linear, label='Linearer Teil')
plt.plot(x, y_power, label='Potenzfunktion')
plt.plot(X, linear(X, phi), 'ro') # Schnittpunkt
plt.xlabel('x')
plt.ylabel('y')
plt.title('L√∂sung der gegebenen Aufgabe')
plt.legend()
plt.grid(True)
plt.show()
```

Berechnen und zeichnen Sie die B√©zierkurve mit den
Bernsteinkoeffizienten f√ºr die B√©zier-Kontrollpunkte
p0 = (2, 2)t, p1 = (3, 5)t, p2 = (5,4)t und p3 = (6,3)t
(Drei Nachkommastellen)

```python
import numpy as np
import matplotlib.pyplot as plt

# Definieren Sie die Kontrollpunkte
p0 = np.array([2, 2])
p1 = np.array([3, 5])
p2 = np.array([5, 4])
p3 = np.array([6, 3])

# Definieren Sie die Funktion f√ºr die Berechnung der B√©zierkurve
def bezier_curve(t, p0, p1, p2, p3):
    return (1-t)**3 * p0 + 3*(1-t)**2 * t * p1 + 3*(1-t) * t**2 * p2 + t**3 * p3

# Erzeugen Sie eine Reihe von t-Werten
t = np.linspace(0, 1, num=1000)

# Berechnen Sie die Punkte auf der B√©zierkurve
points = np.array([bezier_curve(t_val, p0, p1, p2, p3) for t_val in t])

# Zeichnen Sie die B√©zierkurve
plt.figure(figsize=(6, 6))
plt.plot(points[:, 0], points[:, 1], 'b-', linewidth=2)
plt.plot([p0[0], p1[0], p2[0], p3[0]], [p0[1], p1[1], p2[1], p3[1]], 'ro-')
plt.title('B√©zierkurve')
plt.grid(True)
plt.show()
```

$M_{R(30^\circ)} = \begin{bmatrix} \cos 30^\circ & -\sin 30^\circ & 0 \\ \sin 30^\circ & \cos 30^\circ & 0 \\ 0 & 0 & 1 \end{bmatrix} =
\begin{bmatrix} \frac{\sqrt{3}}{2} & -\frac{1}{2} & 0 \\  \frac{1}{2} & \frac{\sqrt{3}}{2} & 0 \\ 0 & 0 & 1 \end{bmatrix}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coordinates of the triangle A(1,1), B(4,3), C(5,2)
triangle = np.array([[1, 4, 5], [1, 3, 2], [1, 1, 1]])

# Define the rotation matrix for 30 degrees
angle = np.radians(30)
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle), 0],
                            [np.sin(angle), np.cos(angle), 0],
                            [0, 0, 1]])

# Perform the rotation in homogeneous coordinates
rotated_triangle = np.dot(rotation_matrix, triangle)


x_limits = [-2, 5]
y_limits = [-2, 6]

# Display the original and rotated triangles
plt.figure(figsize=(8, 4))

# Plot the original triangle
plt.subplot(1, 2, 1)
plt.plot(triangle[0, [0, 1, 2, 0]], triangle[1, [0, 1, 2, 0]], label='Original')
plt.title('Originale Dreieck')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.grid(True)
plt.legend()

# Plot the rotated triangle
plt.subplot(1, 2, 2)
plt.plot(rotated_triangle[0, [0, 1, 2, 0]], rotated_triangle[1, [0, 1, 2, 0]], label='Rotiert')
plt.title('Rotiertes Dreieck (30 Grad um Koordinatenursprung)')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Display the rotated triangle matrix with two decimal places
np.set_printoptions(precision=2, suppress=True)
print("Rotiertes Dreieck Matrix:")
print(rotated_triangle)
```

$M_{T(\overrightarrow{P_R})}= \begin{bmatrix} 1 & 0 & -1 \\ 0 & 1 & -1 \\ 0 & 0 & 1 \end{bmatrix}$  

$M_{T(\overrightarrow{-P_R})}= \begin{bmatrix} 1 & 0 & 1 \\ 0 & 1 & 1 \\ 0 & 0 & 1 \end{bmatrix}$


Rotationswinkel

Rotationsmatrix  
$M_{R(30^\circ)} = \begin{bmatrix} \cos 30^\circ & -\sin 30^\circ & 0 \\ \sin 30^\circ & \cos 30^\circ & 0 \\ 0 & 0 & 1 \end{bmatrix} =
\begin{bmatrix} \frac{\sqrt{3}}{2} & -\frac{1}{2} & 0 \\  \frac{1}{2} & \frac{\sqrt{3}}{2} & 0 \\ 0 & 0 & 1 \end{bmatrix}$ 

$M_{P}'' = M_{T(\overrightarrow{P_R})} \cdot M_{R(30^\circ)} \cdot M_{T(\overrightarrow{-P_R})} \cdot M_P \approx \begin{bmatrix} -0.27 & 1.33 & 2.7 \\ 1.73 & 4.96 & 4.6 \\ 1 & 1 & 1 \end{bmatrix}$


```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_around_point(point, angle):
    translation_matrix = np.array([[1, 0, -point[0]],
                                   [0, 1, -point[1]],
                                   [0, 0, 1]])

    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle), 0],
                                [np.sin(angle), np.cos(angle), 0],
                                [0, 0, 1]])

    inverse_translation_matrix = np.array([[1, 0, point[0]],
                                           [0, 1, point[1]],
                                           [0, 0, 1]])

    return np.dot(inverse_translation_matrix, np.dot(rotation_matrix, translation_matrix))

triangle = np.array([[1, 4, 5], [1, 3, 2], [1, 1, 1]])
rotation_point = np.array([-1, -1])
angle = np.radians(30)

x_limits = [-2, 5]
y_limits = [-2, 6]

rotated_triangle = rotate_around_point(rotation_point, angle) @ triangle

plt.figure(figsize=(8, 4))

# Plot the original triangle
plt.subplot(1, 2, 1)
plt.plot(triangle[0, [0, 1, 2, 0]], triangle[1, [0, 1, 2, 0]], label='Original')
plt.scatter(*rotation_point, color='red', label='Rotations Punkt')
plt.title('Originale Dreieck')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.grid(True)
plt.legend()

# Plot the rotated triangle
plt.subplot(1, 2, 2)
plt.plot(rotated_triangle[0, [0, 1, 2, 0]], rotated_triangle[1, [0, 1, 2, 0]],color='green', label='Rotiert')
plt.scatter(*rotation_point, color='red', label='Rotations Punkt')
plt.title('Rotierte Dreieck (30 Grad)')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

np.set_printoptions(precision=2, suppress=True)
print("Rotiertes Dreieck Matrix:")
print(rotated_triangle)
```
